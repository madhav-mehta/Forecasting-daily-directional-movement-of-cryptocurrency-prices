# Preprocessing

```python
import pandas as pd
import os
```

## Loading Financial Data

```python
btc=pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Financial Data\Historical Data\Bitcoin_hist_24_25.csv',sep=';')
eth=pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Financial Data\Historical Data\Ethereum_hist_24_25.csv',sep=';')
xrp=pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Financial Data\Historical Data\XRP_hist_24_25.csv',sep=';')
```

```python
btc.head()
```

Extracting Date, dropping redundant time columns

```python
btc['date'] = pd.to_datetime(btc['timestamp']).dt.date
btc.drop(columns=['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'timestamp'], inplace=True)

all_columns = btc.columns.tolist()
all_columns.remove('date')
all_columns.insert(0, 'date')
btc = btc[all_columns]
```

```python
eth['date'] = pd.to_datetime(eth['timestamp']).dt.date
eth.drop(columns=['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'timestamp'], inplace=True)

all_columns = eth.columns.tolist()
all_columns.remove('date')
all_columns.insert(0, 'date')
eth = eth[all_columns]
```

```python
xrp['date'] = pd.to_datetime(xrp['timestamp']).dt.date
xrp.drop(columns=['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'timestamp'], inplace=True)

all_columns = xrp.columns.tolist()
all_columns.remove('date')
all_columns.insert(0, 'date')
xrp = xrp[all_columns]
```

## Combining Datasets

### Combining BTC Files

```python
df_crypto = btc
df_sentiment = pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Sentiment Scores\sentiment_scores_BTC_24_25.csv')
df_trends = pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Google Trends\google_trends_BTC_24_25.csv')
df_gold = pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Financial Data\Assets\Gold_data_24_25.csv')
df_sp500 = pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Financial Data\Assets\SP500_data_24_25.csv')
df_nasdaq = pd.read_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Financial Data\Assets\NASDAQ100_data_24_25.csv')
```

```python
df_crypto['date'] = pd.to_datetime(df_crypto['date'])
df_sentiment['date'] = pd.to_datetime(df_sentiment['date'])
df_trends['date'] = pd.to_datetime(df_trends['date'])
df_gold['date'] = pd.to_datetime(df_gold['date'])
df_sp500['date'] = pd.to_datetime(df_sp500['date'])
df_nasdaq['date'] = pd.to_datetime(df_nasdaq['date'])
```

```python
master_df = pd.merge(df_crypto, df_sentiment[['date', 'daily_weighted_sentiment', 'comment_volume']], on='date', how='left')
master_df = pd.merge(master_df, df_trends, on='date', how='left')
master_df = pd.merge(master_df, df_gold[['date', 'gold_spot']], on='date', how='left')
master_df = pd.merge(master_df, df_sp500[['date', 'gspc_spot']], on='date', how='left')
master_df = pd.merge(master_df, df_nasdaq[['date', 'ndx_spot']], on='date', how='left')
```

```python
master_df.head()
```

Forward Filling

```python
master_df.sort_values(by='date', inplace=True)

columns_to_fill = ['gold_spot', 'gspc_spot', 'ndx_spot', 'daily_weighted_sentiment', 'comment_volume', 'BTC_trends']
master_df[columns_to_fill] = master_df[columns_to_fill].ffill()
master_df['BTC_trends'].fillna(36, inplace=True)
```

```python
master_df.head(-1)
```

Checking for missing values

```python
master_df.isnull().sum()
```

```python
master_df.to_csv(r'C:\Users\madha\Desktop\Dissertation\Data\Master Data\Merged\BTC_merged.csv')
```
